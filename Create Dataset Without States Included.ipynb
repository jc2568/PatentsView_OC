```python
import pandas as pd
import time
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv("../csvResults/dfMergedFullDataSet.csv")

display(df.info(),df.head())
```

```python
df1=pd.read_csv("../csvResults/dfMerged_0_274491.csv")

display(df1.info(),df1.head())
```

```python
df1.dropna(subset=['incorporation_date'],inplace=True)

display(df1.info(),df1.head())
```

```python
### changed from ID to assignee_id

print("The number of in unique assignee IDs in the input file is:", df['assignee_id'].nunique())
print("The number of in unique assignee IDs returned in the output file is:", df1['assignee_id'].nunique())
print(f"The percent of returned assignees is: {(df1['assignee_id'].nunique())/df['assignee_id'].nunique():.2%}")
```

```python
noMatch=df.merge(df1,on=['ID'],how='outer',indicator=True)

noMatch.loc[noMatch['_merge']=='left_only']
```

```python
noMatch1=noMatch.loc[noMatch['_merge']=='left_only'].iloc[:,0:1]

noMatch1
```

```python
noMatch2=noMatch1.merge(df,on=['ID'],how='inner')

noMatch2
```

```python
noMatch3=noMatch2.merge(df1,on=['assignee_id'],how='outer',indicator=True)

noMatch4=noMatch3.loc[noMatch3['_merge']=='left_only'].iloc[:,0:7].sort_values(by=['ID_x'])
noMatch4.rename(columns={'ID_x':'ID','state_x':'state'},inplace=True)

noMatch4
```

```python
# noMatch4[noMatch4.duplicated(subset=['assignee_id','organization'],keep='first')]
```

```python
noMatch5=noMatch4.drop_duplicates(subset=['assignee_id','organization'],
                                  keep='first').sort_values(by=['ID']).iloc[:,[1,3,6]].reset_index(drop=True)

display(noMatch5.info(),noMatch5.head())
```

```python
noMatch6=noMatch5.sample(n=1000, random_state=42)#.to_csv("../csvResults/subsetToRerun_1000Records_noStates_Aug_3_2022.csv",index=False)
```

```python
noMatch7=noMatch5.merge(noMatch6,on=['assignee_id','organization','dateOfFirstPat'],how='outer',indicator=True)

display(noMatch7.info(),noMatch7.head())
```

```python
noMatch7.loc[noMatch7['_merge']=='left_only'].iloc[:,0:3].to_csv("../csvResults/subsetToRerun_82614Records_noStates_Aug_9_2022.csv",index=False)
```
